# Análisis del Sistema y Plan de Testing

Este notebook analiza el sistema completo y proporciona tests para identificar problemas.

```python
import os, sys, time
from pathlib import Path
repo_root = Path.cwd()
if not (repo_root / "python" / "ascii_stream_engine").exists():
    repo_root = Path.cwd().parent.parent
sys.path.insert(0, str(repo_root / "python"))
sys.path.insert(0, str(repo_root / "cpp" / "build"))
os.environ["ONNX_MODELS_DIR"] = str(repo_root / "onnx_models" / "mediapipe")
print(f"✅ Configurado: {repo_root}")
```

```python
from ascii_stream_engine.presentation.notebook_api import build_engine_for_notebook, build_general_control_panel
engine = build_engine_for_notebook(0)
panel = build_general_control_panel(engine)
print("✅ Motor y panel creados")
```

## Test 1: Verificar Control de FPS

```python
config = engine.get_config()
print(f"FPS inicial: {config.fps}")
engine.update_config(fps=15)
config_after = engine.get_config()
print(f"FPS después: {config_after.fps}")
print("✅ FPS se actualiza" if config_after.fps == 15 else "❌ FPS NO se actualiza")
```

## Test 2: Verificar Filtros con IA

```python
renderer = engine.get_renderer()
print(f"Renderer: {type(renderer).__name__}")
print(f"Filtros: {len(engine.filters)}")
print("\nOrden: Captura → Análisis → Filtrado → Renderizado (overlay)")
```

## Test 3: Medir Rendimiento

```python
if engine.is_running:
    time.sleep(3)
    metrics = engine.metrics.get_summary()
    print(f"FPS real: {metrics.get('fps', 0):.2f}")
    print(f"FPS objetivo: {engine.get_config().fps}")
else:
    print("⚠️  Motor no corriendo. Inicia desde el panel.")
```
